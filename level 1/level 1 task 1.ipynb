```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, norm
df = pd.read_csv("toy_dataset.csv")  
column = "Age"  
data = df[column]
def plot_pmf_binomial(n=100, p=0.5):
    x = np.arange(n + 1)
    pmf = binom.pmf(x, n, p)
    plt.figure(figsize=(8, 5))
    plt.vlines(x, 0, pmf, colors='b', lw=2)
    plt.scatter(x, pmf, color='b', label="Binomial PMF")
    plt.xlabel("x")
    plt.ylabel("PMF")
    plt.title(f"PMF of a binomial distribution (n={n}, p={p})")
    plt.legend()
    plt.show()
plot_pmf_binomial()
```

```python
def plot_pdf_normal(mean=0, std=1):
    x = np.linspace(-3, 3, 1000)
    pdf = norm.pdf(x, mean, std)
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, label="Normal Distribution")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.title(f"Normal Distribution with mean = {mean} and std = {std}")
    plt.legend()
    plt.show()
plot_pdf_normal()
```

```python
def plot_pdf_cdf_combined(mean=0, std=1):
    x = np.linspace(-2, 2, 1000)
    pdf = norm.pdf(x, mean, std)
    cdf = norm.cdf(x, mean, std)
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, label="Probability Density Function (PDF)", color='b')
    plt.plot(x, cdf, label="Cumulative Distribution Function (CDF)", color='r')
    plt.xlabel("x")
    plt.ylabel("Probability")
    plt.title("Continuous Normal Distributions")
    plt.legend()
    plt.show()
plot_pdf_cdf_combined()
```
