```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(42)
data = pd.DataFrame({
    "Variable_X": np.random.rand(100),
    "Variable_Y": np.random.rand(100)
})
cov_matrix = data.cov()
covariance = cov_matrix.loc["Variable_X", "Variable_Y"]
print("Covariance:\n", covariance)
correlation_matrix = data.corr(method="pearson")
correlation = correlation_matrix.loc["Variable_X", "Variable_Y"]
print("\nPearson's Correlation Coefficient:\n", correlation)
plt.figure(figsize=(6, 4))
sns.scatterplot(x=data["Variable_X"], y=data["Variable_Y"], color='b')
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.title("Scatter Plot of Variable X vs Variable Y")
plt.show()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
