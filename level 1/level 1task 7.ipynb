```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
original_data = np.random.randn(20)  
num_bootstrap_samples = 1000
bootstrap_means = []
for _ in range(num_bootstrap_samples):
    sample = np.random.choice(original_data, size=len(original_data), replace=True)
    bootstrap_means.append(np.mean(sample))
bootstrap_means = pd.Series(bootstrap_means)
plt.figure(figsize=(8, 5))
plt.hist(bootstrap_means, bins=30, density=True, alpha=0.6, color='b')
plt.axvline(np.mean(original_data), color='r', linestyle='dashed', linewidth=2, label='Original Mean')
plt.axvline(np.mean(bootstrap_means), color='g', linestyle='dashed', linewidth=2, label='Bootstrap Mean')
plt.xlabel("Mean Values")
plt.ylabel("Density")
plt.title("Bootstrap Sampling Distribution")
plt.legend()
plt.show()
print("Original Mean:", np.mean(original_data))
print("Bootstrap Mean:", np.mean(bootstrap_means))
print("Bootstrap Standard Error:", np.std(bootstrap_means))
```
