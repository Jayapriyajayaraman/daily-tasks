```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, poisson, norm, lognorm
def plot_uniform():
    data = np.random.uniform(0, 1, 10000)
    plt.figure(figsize=(10, 4))
    sns.histplot(data, bins=30, kde=True, color='b', alpha=0.5)
    plt.title("Random/Uniform Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()

plot_uniform()
```

```python
def plot_normal():
    x = np.linspace(-3, 3, 1000)
    pdf = norm.pdf(x, 0, 1)
    plt.figure(figsize=(10, 4))
    plt.plot(x, pdf, label="Normal PDF", color='b')
    plt.title("Normal Distribution with mean = 0 and std = 1")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

plot_normal()
```

```python
def plot_binomial():
    n, p = 10, 0.5
    x = np.arange(0, n+1)
    pmf = binom.pmf(x, n, p)
    plt.figure(figsize=(10, 4))
    plt.bar(x, pmf, color=['blue', 'orange'], alpha=0.7)
    plt.title(f"Bernoulli Distribution: p = {p}, n = {n}")
    plt.xlabel("Outcome")
    plt.ylabel("Probability")
    plt.show()

plot_binomial()
```

```python
def plot_poisson():
    x = np.arange(0, 20)
    pmf = poisson.pmf(x, 5)
    plt.figure(figsize=(10, 4))
    plt.scatter(x, pmf, color='b', label="Poisson PMF")
    plt.plot(x, pmf, linestyle='dotted', color='b')
    plt.title("Poisson Distribution with lambda=5")
    plt.xlabel("Occurrences")
    plt.ylabel("Probability")
    plt.legend()
    plt.show()

plot_poisson()
```

```python
def plot_lognormal():
    x = np.linspace(0.1, 5, 1000)
    pdf = lognorm.pdf(x, s=0.9)
    cdf = lognorm.cdf(x, s=0.9)
    plt.figure(figsize=(10, 4))
    plt.plot(x, pdf, label="Log-normal PDF", color='b')
    plt.plot(x, cdf, label="Log-normal CDF", color='orange')
    plt.title("Visualization of Log-normal PDF and CDF")
    plt.xlabel("Value")
    plt.ylabel("Probability")
    plt.legend()
    plt.show()

plot_lognormal()
```
