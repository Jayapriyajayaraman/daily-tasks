```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
income_data = np.random.normal(loc=100000, scale=20000, size=100000)  
Q1 = np.percentile(income_data, 25)  
Q3 = np.percentile(income_data, 75) 
IQR = Q3 - Q1  
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = income_data[(income_data < lower_bound) | (income_data > upper_bound)]
normal_data = income_data[(income_data >= lower_bound) & (income_data <= upper_bound)]
plt.figure(figsize=(10, 5))
plt.hist(normal_data, bins=100, alpha=0.8, label="Income distribution", color="steelblue")
plt.hist(outliers, bins=100, alpha=0.8, label="Outliers", color="orangered")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.title("Outlier detection a la Tukey's method")
plt.legend()
plt.show()
```
